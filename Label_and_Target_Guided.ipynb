##Label And Ordinal Encoding

```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# create a sample dataframe with a categorical variable
df = pd.DataFrame({
    'color': ['red','blue','green','green','red','blue']
})
```

```python
df
```

```python
## create an instance of labelencoder
encoder=LabelEncoder()
```

```python
encoder.fit_transform(df['color'])
```

```python
## Ordinal Encoding
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

#create a  ssamplr dataframe with an ordinal variable
df = pd.DataFrame({
    'size': ['small','medium','large','medium','small','large']
})
```

```python
## create an instance of the OrdinalEncoder class and fit_transform
encoder = OrdinalEncoder(categories=[['small','medium','large']])
```

```python
encoder.fit_transform(df[['size']])
```

## Target Guided Ordinal Encoding

```python
import pandas as pd
# create a sample datafrme with a categorical variable and a target variable
df = pd.DataFrame({
    'city': ['New york','London','Paris','Tokyo','New York','Paris'],
    'price': [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

```python
df.head()
```

```python
## calculate the mean price for each city
mean_price = df.groupby('city')['price'].mean().to_dict()
mean_price
```

```python
## replace each city with its mean price
df['city_encoded'] = df['city'].map(mean_price)
```
